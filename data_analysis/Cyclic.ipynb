```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
sns.set_style('whitegrid')
```

```python
filepath = '../avida_files/clock/data/cyclic/zero/'
```

```python
columns = ['update','day','night','']

df_cyc_res = pd.read_csv(filepath+"resource.dat",
                         delimiter=' ', skiprows=8, header=None)

df_cyc_res.columns = columns
df_cyc_res = df_cyc_res.dropna(axis=1)
```

```python
df_cyc_res.plot('update',['day','night'], kind='line', ls='-', lw=1,
                color=['r','b'])
plt.ylim(-1,11)
plt.xlim(0,1000)
plt.title('Cyclic Env')
```

```python
columns = ['update','merit','gestation_time','fitness','repro_rate',
           '(deprecated)_size','copied_size','executed_size',
           '(deprecated)_abundance','proportion_of_organisms_that_gave_birth_in_this_update',
           'proportion_of_breed_true_organisms','(deprecated)_genotype_depth',
           'generation','neutral_metric','lineage_label',
           'true_replication_rate_(based_on_births/update_time-averaged)','']

df_cyc_ave = pd.read_table(filepath+"average.dat",
                           delimiter=' ', skiprows=19, header=None)

df_cyc_ave.columns = columns
df_cyc_ave = df_cyc_ave.dropna(axis=1)
```

```python
plt.plot(df_cyc_ave['update'], df_cyc_ave['fitness'], color='crimson')
plt.fill_between(df_cyc_res['update'], 0, df_cyc_res['night']*0.35e43, alpha=0.2, color='grey')
plt.xlim(295000,298000)
plt.ylim(0.99e42, 3.5e43);
```

```python
columns = ['update','not','nand','and','ornot','or','andnot','nor','xor','equals','']

df_cyc_reac = pd.read_table(filepath+"reactions_exe.dat",
                            delimiter=' ', skiprows=16, header=None)

df_cyc_reac.columns = columns
df_cyc_reac = df_cyc_reac.dropna(axis=1)
```

```python
df_cyc_reac.plot('update', ['not','nand','and','ornot','or',
                            'andnot','nor','xor','equals'])
```

```python
df_cyc_reac.plot('update', ['not','nand','and','ornot','or',
                            'andnot','nor','xor','equals'],
                 color=['mistyrose','lavender','lightsalmon','lightsteelblue','salmon',
                        'cornflowerblue','crimson','navy','firebrick'])
plt.fill_between(df_cyc_res['update'], 0, df_cyc_res['night']*1600, alpha=0.2, color='grey')
plt.xlim(295000,298000)
plt.legend(bbox_to_anchor=(1.25, 1.0))

plt.savefig('Cyclic.pdf', bbox_inches='tight')
```
