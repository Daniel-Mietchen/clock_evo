```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
sns.set_style('whitegrid')
```

```python
columns = ['update','day','night','']

df_cyc0_res = pd.read_csv("../avida_files/clock/data/cyclic/zero/resource.dat",
                         delimiter=' ', skiprows=8, header=None)

df_cyc0_res.columns = columns
df_cyc0_res = df_cyc0_res.dropna(axis=1)

############################

df_cyc_res = pd.read_csv("../avida_files/clock/data/cyclic/not_zero/resource.dat",
                         delimiter=' ', skiprows=8, header=None)

df_cyc_res.columns = columns
df_cyc_res = df_cyc_res.dropna(axis=1)

############################

df_cp_res = pd.read_csv("../avida_files/clock/data/punishment/value/resource.dat",
                         delimiter=' ', skiprows=8, header=None)

df_cp_res.columns = columns
df_cp_res = df_cp_res.dropna(axis=1)

############################

df_pun_res = pd.read_csv("../avida_files/clock/data/punishment/mult/resource.dat",
                         delimiter=' ', skiprows=8, header=None)

df_pun_res.columns = columns
df_pun_res = df_pun_res.dropna(axis=1)
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, sharex='col', sharey='row')
ax1.step(df_cyc0_res['update'], df_cyc0_res[['day','night']])
ax1.set_title('Cyclic 0/10')
ax1.set_ylim(-1,11)
ax1.set_xlim(0,100)

ax2.step(df_cyc_res['update'], df_cyc_res[['day','night']])
ax2.set_title('Cyclic 1/10')
ax2.set_ylim(-1,11)
ax2.set_xlim(0,100)
ax2.legend(['day','night'], bbox_to_anchor=(1.45, 1.0))

ax3.plot(df_cp_res['update'], df_cp_res[['day','night']])
ax3.set_title('Punishment_Value')
ax3.set_ylim(-1,11)
ax3.set_xlim(0,100)

ax4.plot(df_pun_res['update'], df_pun_res[['day','night']])
ax4.set_title('Punishment_Mult')
ax4.set_ylim(-1,11)
ax4.set_xlim(0,100);
```

```python
columns = ['update','merit','gestation_time','fitness','repro_rate',
           '(deprecated)_size','copied_size','executed_size',
           '(deprecated)_abundance','proportion_of_organisms_that_gave_birth_in_this_update',
           'proportion_of_breed_true_organisms','(deprecated)_genotype_depth',
           'generation','neutral_metric','lineage_label',
           'true_replication_rate_(based_on_births/update_time-averaged)','']


df_cyc0_ave = pd.read_csv("../avida_files/clock/data/cyclic/zero/average.dat",
                         delimiter=' ', skiprows=19, header=None)

df_cyc0_ave.columns = columns
df_cyc0_ave = df_cyc0_ave.dropna(axis=1)

############################

df_cyc_ave = pd.read_csv("../avida_files/clock/data/cyclic/not_zero/average.dat",
                         delimiter=' ', skiprows=19, header=None)

df_cyc_ave.columns = columns
df_cyc_ave = df_cyc_ave.dropna(axis=1)

############################

df_cp_ave = pd.read_csv("../avida_files/clock/data/punishment/value/average.dat",
                         delimiter=' ', skiprows=19, header=None)

df_cp_ave.columns = columns
df_cp_ave = df_cp_ave.dropna(axis=1)

############################

df_pun_ave = pd.read_csv("../avida_files/clock/data/punishment/mult/average.dat",
                         delimiter=' ', skiprows=19, header=None)

df_pun_ave.columns = columns
df_pun_ave = df_pun_ave.dropna(axis=1)
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, sharex='col', sharey='row')
ax1.plot(df_cyc0_ave['update'], df_cyc0_ave['fitness'])
ax1.set_title('Cyclic 0/10')
ax1.set_yscale("log")

ax2.plot(df_cyc_ave['update'], df_cyc_ave['fitness'])
ax2.set_title('Cyclic 1/10')

ax3.plot(df_cp_ave['update'], df_cp_ave['fitness'])
ax3.set_title('Punishment_Value')
ax3.set_yscale("log")
ax3.set_xticklabels(['0','5k','10k','15k','20k','25k'])

ax4.plot(df_pun_ave['update'], df_pun_ave['fitness'])
ax4.set_title('Punishment_Mult')
ax4.set_xticklabels(['0','5k','10k','15k','20k','25k']);
```

```python
columns = ['update','not','nand','and','ornot','or','andnot','nor','xor','equals','']

df_cyc0_reac = pd.read_table("../avida_files/clock/data/cyclic/zero/reactions_exe.dat",
                            delimiter=' ', skiprows=16, header=None)

df_cyc0_reac.columns = columns
df_cyc0_reac = df_cyc0_reac.dropna(axis=1)

############################

df_cyc_reac = pd.read_table("../avida_files/clock/data/cyclic/not_zero/reactions_exe.dat",
                            delimiter=' ', skiprows=16, header=None)

df_cyc_reac.columns = columns
df_cyc_reac = df_cyc_reac.dropna(axis=1)

############################

df_cp_reac = pd.read_table("../avida_files/clock/data/punishment/value/reactions_exe.dat",
                            delimiter=' ', skiprows=16, header=None)

df_cp_reac.columns = columns
df_cp_reac = df_cp_reac.dropna(axis=1)

############################

df_pun_reac = pd.read_table("../avida_files/clock/data/punishment/mult/reactions_exe.dat",
                            delimiter=' ', skiprows=16, header=None)

df_pun_reac.columns = columns
df_pun_reac = df_pun_reac.dropna(axis=1)
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, sharex='col', sharey='row')
ax1.plot(df_cyc0_reac['update'], df_cyc0_reac[['not','nand','and','ornot','or',
                                               'andnot','nor','xor','equals']])
ax1.set_title('Cyclic 0/10')

ax2.plot(df_cyc_reac['update'], df_cyc_reac[['not','nand','and','ornot','or',
                                              'andnot','nor','xor','equals']])
ax2.set_title('Cyclic 1/10')
ax2.legend(['not','nand','and','ornot','or','andnot','nor','xor','equals'], bbox_to_anchor=(1.5, 1.1))

ax3.plot(df_cp_reac['update'], df_cp_reac[['not','nand','and','ornot','or',
                                           'andnot','nor','xor','equals']])
ax3.set_title('Punishment_Value')
ax3.set_xticklabels(['0','5k','10k','15k','20k','25k'])

ax4.plot(df_pun_reac['update'], df_pun_reac[['not','nand','and','ornot','or',
                                             'andnot','nor','xor','equals']])
ax4.set_title('Punishment_Mult')
ax4.set_xticklabels(['0','5k','10k','15k','20k','25k']);
```
