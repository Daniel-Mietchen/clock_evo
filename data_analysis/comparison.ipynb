```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
sns.set_style('whitegrid')
```

```python
columns = ['update','day','night','']

df_cyc_res = pd.read_csv("../avida_files/clock/data/cyclic/resource.dat",
                         delimiter=' ', skiprows=8, header=None)

df_cyc_res.columns = columns
df_cyc_res = df_cyc_res.dropna(axis=1)

######################

df_pun_res = pd.read_csv("../avida_files/clock/data/punishment/resource.dat",
                         delimiter=' ', skiprows=8, header=None)

df_pun_res.columns = columns
df_pun_res = df_pun_res.dropna(axis=1)

######################

df_cp_res = pd.read_csv("../avida_files/clock/data/cyclic+punishment/resource.dat",
                         delimiter=' ', skiprows=8, header=None)

df_cp_res.columns = columns
df_cp_res = df_cp_res.dropna(axis=1)
```

```python
df_cyc_res.plot('update',['day','night'], kind='line', ls='-', lw=1,
                color=['r','b'])
plt.ylim(-10,110)
plt.title('Cyclic Env')

df_pun_res.plot('update',['day','night'], kind='line', ls='-', lw=1,
                color=['r','b'])
plt.ylim(-10,110)
plt.title('Punishment Env')

df_cp_res.plot('update',['day','night'], kind='line', ls='-', lw=1,
               color=['r','b'])
plt.ylim(-10,110)
plt.title('Cyclic+Punishment Env');
```
